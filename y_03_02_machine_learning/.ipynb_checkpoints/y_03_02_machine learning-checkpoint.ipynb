Распространённая проблема фитнес-клубов и других сервисов — отток клиентов.<br>  
Индикаторы оттока зависят от специфики отрасли. Для фитнес-центра можно считать, что клиент попал в отток, если за последний месяц ни разу не посетил спортзал. Конечно, не исключено, что он уехал на Бали и по приезде обязательно продолжит ходить на фитнес. Однако чаще бывает наоборот. Если клиент начал новую жизнь с понедельника, немного походил в спортзал, а потом пропал — скорее всего, он не вернётся.<br> 
Чтобы бороться с оттоком, отдел по работе с клиентами «Культуриста-датасаентиста» перевёл в электронный вид множество клиентских анкет. Задача — провести анализ и подготовить план действий по удержанию клиентов.<br>

•	прогнозировать вероятность оттока (на уровне следующего месяца) для каждого клиента;<br>
•	сформировать типичные портреты клиентов: выделить несколько наиболее ярких групп и охарактеризовать их основные свойства;<br>
•	проанализировать основные признаки, наиболее сильно влияющие на отток;<br>
•	сформулировать основные выводы и разработать рекомендации по повышению качества работы с клиентами:<br> 
	1) выделить целевые группы клиентов;<br>
	2) предложить меры по снижению оттока;<br>
	3) определить другие особенности взаимодействия с клиентами.<br>

Сгруппируем клиентов фитнес-центра по признаку "ушел / остался" и:<br>
- выведем средние значения признаков в двух группах;<br>
- построим столбчатые гистограммы и распределения признаков для тех, кто ушёл (отток) и тех, кто остался (не попали в отток);<br>
- построим матрицу корреляций и отобразим её.<br>

```python
# Собираем датафрейм по фильтру "остался" (сhurn = 0):
df_сhurn_0 = df.query('churn == 0')
сhurn_0 = df_сhurn_0.shape[0]
#print(сhurn_0)
#print('Датафрейм по фильтру "остался" (сhurn = 0):')
#display(df_сhurn_0.head())
#print(df_сhurn_0.info())

# Собираем датафрейм по фильтру "ушел" (сhurn = 1):
df_сhurn_1 = df.query('churn == 1')
сhurn_1 = df_сhurn_1.shape[0]
#print(сhurn_1)
#print('Датафрейм по фильтру "ушел" (сhurn = 1):')
#display(df_сhurn_1.head())
#print(df_сhurn_1.info())

# Собираем датафрейм из средних значений по фильтру "ушел" и "остался":
print('Датафрейм из средних значений по фильтру "ушел" (сhurn = 1) и "остался" (сhurn = 0):')
df_сhurn_0_mean = df_сhurn_0.describe().reset_index()
df_сhurn_0_mean.loc[1, 'index'] = сhurn_0
df_сhurn_0_mean.rename(columns = {'index' : 'clients_count'}, inplace = True)
df_сhurn_0_mean = df_сhurn_0_mean.loc[1, :]
df_сhurn_1_mean = df_сhurn_1.describe().reset_index()
df_сhurn_1_mean.loc[1, 'index'] = сhurn_1
df_сhurn_1_mean.rename(columns = {'index' : 'clients_count'}, inplace = True)
df_сhurn_1_mean = df_сhurn_1_mean.loc[1, :]
df_mean = pd.concat([df_сhurn_0_mean, df_сhurn_1_mean], axis = 1).T.reset_index().drop(labels = 'index', axis = 1)
df_mean.loc[0, 'churn'] = 'клиент остался'
df_mean.loc[1, 'churn'] = 'клиент ушел'
display(df_mean)
```

```python
# Строим столбчатые гистограммы и распределения признаков по фильтру "ушел" и "остался":
print('Столбчатые гистограммы по фильтру "ушел" (сhurn = 1) и "остался" (сhurn = 0):')
for col in df_mean.drop('churn', axis = 1).columns:
    fig, ax = plt.subplots(figsize = (12, 4))
    sns.barplot(x = df_mean['churn'], y = df_mean[col])
    plt.show();   
```

```python
# Строим и отрисовываем матрицу корреляций по исходному датафрейму:
print('Матрица корреляций по исходному датафрейму:')
# вычисляем матрицу корреляций
#cm = df.groupby('сhurn')['сhurn'].corr() 
cm = df.corr()
display(cm)
# рисуем тепловую карту с подписями для матрицы корреляций
fig, ax = plt.subplots(figsize = (14, 14))
sns.heatmap(cm, annot=True, square=True) 
ax.set_ylim(14, 0)  # корректировка "рваных" полей heatmap в последней версии библиотеки
plt.show()
```

Возможные алгоритмы: логистическая регрессия, дерево принятия решений, случайный лес, градиентный бустинг.<br>

Построим модель бинарной классификации клиентов, где целевой признак — факт оттока клиента в следующем месяце:<br>
- Разобъем данные на обучающую и валидационную выборку в пропорции 80/20;<br>
- Обучим модель на обучающей выборке (train-выборке) двумя способами:<br> 
  а) логистической регрессией (реализация алгоритма: sklearn.linear_model.LogisticRegression()),<br>
  б) случайным лесом (реализация алгоритма: sklearn.ensemble.RandomForestClassifier()).<br>
- Оценим метрики "доля правильных ответов" (accuracy), "точность" (precision) и "полнота" (recall) для обеих моделей на валидационной выборке. Сравним по ним модели и определим модель, которая показала себя лучше на основании метрик.

```python
print('Размер датасета для проверки моделей:', df_clean.shape)

# посмотрим, как соотносятся классы набора данных
print('Соотношение классов (клиент остался churn = 0 или клиент ушел churn = 1):\n', df_clean['churn'].value_counts())

# разделим наши данные на признаки (матрица X) и целевую переменную (y)
X = df_clean.drop(columns = ['churn'])
y = df_clean['churn']
print()

# определим функцию, которая будет выводить наши метрики
def print_all_metrics(y_true, y_pred, y_proba, title='Метрики классификации'):
    print(title)
    print('\t1) доля правильных ответов (Accuracy)....: {:.2f}'.format(accuracy_score(y_true, y_pred)))
    print('\t2) точность (Precision)..................: {:.2f}'.format(precision_score(y_true, y_pred)))
    print('\t3) полнота (Recall)......................: {:.2f}'.format(recall_score(y_true, y_pred)))
    print('\t4) баланс между точностью и полнотой (F1): {:.2f}'.format(f1_score(y_true, y_pred)))
    print('\t5) площадь под кривой ошибок (ROC_AUC)...: {:.2f}'.format(roc_auc_score(y_true, y_proba)))

# разделяем модель на обучающую и валидационную выборку в пропорции 80/20
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# стандартизуем данные: создадим объект класса StandardScaler и применим его к обучающей выборке
scaler = StandardScaler()
X_train_st = scaler.fit_transform(X_train)  # обучаем scaler и одновременно трансформируем матрицу для обучающей выборки

# применяем стандартизацию к матрице признаков для тестовой выборки
X_test_st = scaler.transform(X_test)

# обучим StandartScaler на обучающей выборке
scaler = StandardScaler()
scaler.fit(X_train)

# Преобразуем обучающий и валидационные наборы данных
X_train_st = scaler.transform(X_train)
X_test_st = scaler.transform(X_test)

# зададим алгоритм для модели логистической регрессии
lr_model = LogisticRegression(solver = 'liblinear', random_state = 0)
# обучим модель
lr_model.fit(X_train_st, y_train)
# воспользуемся уже обученной моделью, чтобы сделать прогнозы
lr_predictions = lr_model.predict(X_test_st)
lr_probabilities = lr_model.predict_proba(X_test_st)[:, 1]
# выведем все метрики
print_all_metrics(
    y_test,
    lr_predictions,
    lr_probabilities,
    title = 'Метрики для модели логистической регрессии (реализация алгоритма: sklearn.linear_model.LogisticRegression()):',
)
print()

# зададим алгоритм для новой модели на основе алгоритма случайного леса
rf_model = RandomForestClassifier(n_estimators = 100, random_state = 0) 
# обучим модель случайного леса
rf_model.fit(X_train_st, y_train)
# воспользуемся уже обученной моделью, чтобы сделать прогнозы
rf_predictions = rf_model.predict(X_test_st)
rf_probabilities = rf_model.predict_proba(X_test_st)[:, 1]
# выведем все метрики
print_all_metrics(
    y_test,
    rf_predictions,
    rf_probabilities, 
    title='Метрики для модели случайного леса (реализация алгоритма: sklearn.ensemble.RandomForestClassifier()):'
)
```

Оценивая метрики "доля правильных ответов" (accuracy), "точность" (precision) и "полнота" (recall) для обеих моделей на валидационной выборке, заключаем, что модель случайного леса по двум из трех метрик (accuracy и precision) превосходит модель логистической регрессии.<br>
В то же время по дополнительным обобщающим метрикам "баланс между точностью и полнотой" (F1) и "площадь под кривой ошибок" (ROC_AUC) - у модели логистической регрессии показатели чуть лучше.<br>

Поскольку согласно задания приоритетными являются метрики accuracy, precision и recall, то окончательно считаем модель случайного леса лучшей из двух рассмотренных моделей.

## Шаг 4. Кластеризация клиентов
[В начало](#intro)

Проведем кластеризацию объектов (клиентов). Для этого выполним следующие действия:<br>
• Стандартизируем данные.<br>
• Построим матрицу расстояний на стандартизованной матрице признаков и нарисуем дендрограмму. На основании полученного графика предположим, какое количество кластеров можно выделить.<br>
• Обучим модель кластеризации на основании алгоритма K-Means и спрогнозируем кластеры клиентов. Примем число кластеров n=5.<br>
• Оценим средние значения признаков для кластеров.<br>
• Построим распределения признаков для кластеров.<br>
• Для каждого полученного кластера посчитаем долю оттока и определим, какие кластеры склонны к оттоку, а какие — надёжны.<br>

```python
# Определим функцию отрисовки графиков попарных признаков для кластеров
def show_clusters_on_plot(df, x_name, y_name, cluster_name):
    plt.figure(figsize = (5, 5))
    sns.scatterplot(df[x_name], df[y_name], hue = df[cluster_name], palette = 'Paired')
    plt.title('{} vs {}'.format(x_name, y_name))
    plt.show()

# Стандартизируем данные
sc = StandardScaler() # создаём объект класса scaler (нормализатор)
x_sc = sc.fit_transform(df_clean.drop(columns = ['churn'])) # обучаем нормализатор и преобразуем набор данных 
print('Стандартизируем данные.\nНа выходе получим преобразованные таблицы объект-признак для обучающего и валидационного набора.\nИ уже их будем применять для обучения и прогнозирования.\nПосле преобразования нашей исходной таблицы она стала выглядеть вот так:')
print(x_sc[:10])
```

```python
# Матрица расстояний на стандартизованной матрице признаков и отрисовка дендрограммы:

print('Матрица расстояний на стандартизованной матрице признаков:')
# обязательная стандартизация данных перед работой с алгоритмами
sc = StandardScaler()
X_sc = sc.fit_transform(X)
linked = linkage(X_sc, method = 'ward') # в переменной linked сохранена таблица «связок» между объектами
print(linked)
print()

print('Дендрограмма по матрице расстояний на стандартизованной матрице признаков:')
# визуализируем таблицу «связок» между объектами как дендрограмму: 
plt.figure(figsize=(15, 10))  
dendrogram(linked, orientation = 'top')
plt.axhline(y = 60, c = 'grey', lw = 2, linestyle = 'dashed') # проводим черту для дополнительной визуализации ветвей (кластеров)
plt.title('Иерархическая кластеризация для фитнес-центра')
plt.show() 
```

На основании полученного графика возможно предположить, что оптимальное количество кластеров равно 4. Тем не менее, поскольку "фиолетовая" зона на дендрограмме занимает практически половину отображаемой оси, то, вероятно, имеет смысл увеличить количество кластеров. По условию задания окончательно принимаем число кластеров равным 5.

```python
# Обучим модель кластеризации на основании алгоритма K-Means и спрогнозируем кластеры клиентов:

# стандартизируем данные
sc = StandardScaler()
x_sc = sc.fit_transform(df_clean)

# задаём модель k_means с числом кластеров 5 и фиксируем значение random_state
model_churn = KMeans(n_clusters = 5, random_state = 0)

# прогнозируем кластеры для наблюдений (алгоритм присваивает им номера от 0 до 4)
labels = model_churn.fit_predict(x_sc)

# сохраняем метки кластера по каждому клиенту в новом столбце нашего датасета
df_clean['cluster_churn'] = labels

# выводим статистику по средним значениям наших признаков по кластеру
print('Средние значения признаков для кластеров "cluster_churn"\n(модель кластеризации на основании алгоритма K-Means, n_clusters = 5):')
display(df_clean.groupby(['cluster_churn']).mean())
print()
```

Оценивая средние значения признаков для кластеров "cluster_churn" отмечаем, что есть два кластера со "стабильными" клиентами - отток минимальный (кластеры № 1 и 2 с оттоком 0,5% м 0,2% соответственно). Также выделяем два кластера "колеблющихся" клиентов (кластеры № 3 и 4 с оттоком 26,7% м 40,4% соответственно). Кроме того, определена однозначно "рискованная" группа клиентов, которые почти наверняка совсем скоро перестанут посещать фитнес-центр (кластер № 0 с оттоком 99,3%).

```python
# Создадим функцию, которая категоризирует клиентов в зависимости от выбранных кластеров как 
# "А1_стабильный / А2_стабильный / Б1_колеблющийся / Б2_колеблющийся / В_рискованный":   
def get_cluster_churn(row):
    x = row['cluster_churn']
    if x == 2: return 'А1_стабильный'
    elif x == 1: return 'А2_стабильный'
    elif x == 3: return 'Б1_колеблющийся'
    elif x == 4: return 'Б2_колеблющийся'
    else: return 'В_рискованный'   
    
# Создадим столбец client_categoty с категориями из функции get_cluster_churn:
df_clean['client_categoty'] = df_clean.apply(get_cluster_churn, axis=1)

print('Первые  и последние 5 строк анализируемого датасета с присвоенными метками кластера и категориям для каждого клиента и\nотсортированного по этим категориям:')
df_clean = df_clean.sort_values(by = 'client_categoty')
display(df_clean)
```

```python
# Строим столбчатые гистограммы и распределения признаков по фильтру 'client_categoty':
print('Столбчатые гистограммы по фильтру "client_categoty":')
for col in df_clean.drop(['churn', 'cluster_churn', 'client_categoty'], axis = 1).columns:
    fig, ax = plt.subplots(figsize = (12, 4))
    sns.barplot(x = df_clean['client_categoty'], y = df_clean[col])
    plt.show();      

# Постоим графики распределения признаков для кластеров (графики для попарных сочетаний признаков с разметкой по кластерам):
print('Графики для попарных сочетаний признаков с разметкой по кластерам "cluster_churn":')
col_pairs = list(itertools.combinations(df_clean.drop(['cluster_churn', 'client_categoty'], axis = 1).columns, 2))
for pair in col_pairs:
    show_clusters_on_plot(df_clean, pair[0], pair[1], 'client_categoty')
```

Оценивая cтолбчатые гистограммы по фильтру "client_categoty" и графики попарных распределений признаков для кластеров "cluster_churn" в контексте поставленной в задании задачи отмечаем, что ранее сделанные выводы подтверждаются визуализацией: есть два кластера со "стабильными" клиентами - отток минимальный; два кластера "колеблющихся" клиентов и "рискованная" группа, клиенты из которой почти наверняка совсем скоро перестанут посещать фитнес-центр.

```python
# Для каждого полученного кластера считаем долю оттока:
print('Доли оттока для полученных кластеров:')
n_clusters = df_clean['client_categoty'].unique()
client_total = df_clean.shape[0]
#print(n_clusters, client_total)
for i in n_clusters:
    print('---------------------------------------------------------------------------------------------')
    print('Для кластера "', i, '"')
    var_1 = np.array(df_clean.query('client_categoty == @i').groupby('cluster_churn')['churn'].count())
    print('- количество клиентов в кластере....:', var_1, 'или', ((var_1 / client_total) * 100).round(2), '% от всех клиентов фитнес-центра')
    var_2 = np.array((df_clean.query('client_categoty == @i').groupby('cluster_churn')['churn'].mean() * 100).round(2))
    print('- доля оттока клиентов в кластере, %:', var_2)
```

Оценивая полученные значения для кластеров отмечаем, что чуть больше половины клиентов относятся к "стабильным" (58,18%). На долю "колеблющихся" приходится порядка 23,57%. Вероятнее всего фитнес-центр потеряет всех клиентов "рискованной" группы, а это порядка 18,25% от общего количества клиентов.
