## Предобработка данных
[В начало](#intro)

Заменим названия столбцов (приведем их к нижнему регистру).</br>

Переопределим типы данных.</br>
Данные в столбце Year_of_Release (год выпуска) должен быть в формате целого числа, а не вещественного: меняем формат.</br>
Аббревиатуру "tbd" в столбце User_Score (оценка пользователей) расшифровываем как "To Be Determined (будет определено)" и заменяем эту аббревиатуру на пропуски (NaN), после чего меняем формат из строки в вещественное число.

Пропуски в данных.</br>
Имеются пропуски данных в столбцах: name, year_of_release, genre, critic_score, user_score и rating.</br>
Для целей дальнейшего исследования неважно название игр - оставляем столбец name с пропусками.</br>
Для остальных рассматриваемых столбцов пропуски также оставляем, поскольку замена какими-либо значениями может привести к искажению итоговой картины (например, замена медианой пропущенных оценок от критиков может исказить результаты в сравнении по жанрам).</br>

Возможные причины, которые могли привести к пропускам:</br>
* часть игр очень старые и, вероятно, статистика по ним просто отсутствует;
* для другой части игр отдельные поля, вероятно, не заполнялись при составлении базы данных (например, год выхода для "FIFA Soccer 2004" или "LEGO Batman: The Videogame").

Посчитаем суммарные продажи во всех регионах и запишем их в отдельный столбец total_sales и удалим дубликаты (при их наличии).



```python
# Переименовываем столбцы (приводим названия к нижнему регистру):
df.columns = df.columns.str.lower()

# Переопределяем типы данных:
df['year_of_release'] = df['year_of_release'].astype('Int64')
df.loc[(df['user_score'] == "tbd"), 'user_score'] = np.nan
df['user_score'] = df['user_score'].astype('float64')

# Cчитаем суммарные продажи во всех регионах и записываем их в отдельный столбец total_sales:
df['total_sales'] = df[['na_sales','eu_sales','jp_sales', 'other_sales']].sum(axis = 1)

# Удаляем дубликаты:
print(f'Количество строк в датафрейме после предобработки: {df.shape[0]}')
print(f'Количество дубликатов в датафрейме после предобработки: {df.duplicated().sum()}, или {round(df.duplicated().sum() / df.shape[0] * 100)}% от общего количества строк предобработанного датафрейма.')
df = df.drop_duplicates()
print(f'Количество строк в датафрейме после предобработки и удаления дубликатов: {df.shape[0]}')

# Информация о датафреме после предобработки:
print("-----------------------------------------------------------------------------")
print("Первые 5 строк предобработанного датафрейма:")
display(df.head())
print("-----------------------------------------------------------------------------")
print("Строки предобработанного датафрейма, в которых не указано название игры:")
display(df[df['name'].isna()])
print("-----------------------------------------------------------------------------")
print("Общая информация о датафрейме:")
df.info()
print("-----------------------------------------------------------------------------")
print("Разброс значений, максимум, минимум, среднее и медиана в столбцах датафрейма:")
display(df.describe())
print("=============================================================================")
```

Посмотрим, сколько игр выпускалось в разные годы.

```python
# Узнаем количество игр, выпущенных в разные годы:
print('Количество и объемы продаж игр, выпущенных в разные годы:')
#display(df.groupby('year_of_release')['name'].count())
all_games = df.groupby('year_of_release')['name'].count().reset_index()
all_games_sales = df.groupby('year_of_release')['total_sales'].sum().reset_index()
all_games = all_games.merge(all_games_sales, on=['year_of_release'], how='outer')
all_games.rename(columns = {'name' : 'games_count', 'total_sales' : 'games_total_sales'}, inplace = True)
display(all_games)

df.groupby('year_of_release')['name'].count().plot(
                                                   xlabel='Год выпуска', 
                                                   ylabel='Количество выпущенных игр, шт.', 
                                                   figsize=(12, 4), 
                                                   grid=True
                                                   )
plt.title('Количество игр, выпущенных в разные годы')
plt.show()
```

Посмотрим, **как влияют на продажи отзывы пользователей и критиков для популярной платформы PS4**. Построим диаграмму рассеяния и посчитаем корреляцию между отзывами и продажами.

### Составление портрета пользователя каждого региона

Составим портрет пользователя для регионов: Северная Америка (NA), Европейский союз (EU), Япония (JP), остальные регионы (Other).</br>

Для этого определим:
- доли продаж самых популярных платформ (ТОР-5).
- доли самых популярных жанров (ТОР-5).
- влияние рейтинга ESRB на продажи в отдельном регионе.

```python
# Названия ТОР-5 выбранных платформ: 
print('Названия ТОР-5 выбранных платформ:', games_platform_2014_2016_top_names)

# Собираем таблицу для составления портрета пользователей заменяя нетоповые платформы словом "other": 
df_portrets = df_2014_2016.copy()
df_portrets['platform'] = (df_portrets['platform']
                           .where(df_portrets['platform']
                           .isin(games_platform_2014_2016_top_names), 'other'))

# Строим графики продаж платформ:
#     готовим список названий регионов:
regions = ['Северная Америка (NA)', 'Европейский союз (EU)', 'Япония (JP)', 'Остальные регионы (Other)']
#     готовим список названий платформ:
platforms = df_portrets['platform'].unique()
#     готовим список названий столбцов продаж для регионов:
sales = ['na_sales', 'eu_sales', 'jp_sales', 'other_sales']
#     готовим сводные таблицы с данными о продажах по платформам и регионам:     
na = df_portrets.pivot_table(index='platform', values='na_sales', aggfunc='sum').reset_index()
eu = df_portrets.pivot_table(index='platform', values='eu_sales', aggfunc='sum').reset_index()
jp = df_portrets.pivot_table(index='platform', values='jp_sales', aggfunc='sum').reset_index()
other = df_portrets.pivot_table(index='platform', values='other_sales', aggfunc='sum').reset_index()
#     рисуем круговые диаграммы:
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(18, 4))
fig.suptitle('Доля продаж игр для различных платформ по регионам, %', y=1.012, fontsize=16)
axes[0].set(title='Северная Америка (NA)')
axes[0].pie(na['na_sales'], labels=na['platform'], autopct='%1.1f%%', shadow=True)
axes[1].set(title='Европейский союз (EU)')
axes[1].pie(eu['eu_sales'], labels=eu['platform'], autopct='%1.1f%%', shadow=True)
axes[2].set(title='Япония (JP)')
axes[2].pie(jp['jp_sales'], labels=jp['platform'], autopct='%1.1f%%', shadow=True)
axes[3].set(title='Остальные регионы (Other)')
axes[3].pie(other['other_sales'], labels=other['platform'], autopct='%1.1f%%', shadow=True)
plt.show()
#     рисуем гистограммы:
fig, axs = plt.subplots(nrows=1, ncols=4, figsize=(18, 4))
fig.suptitle('Объем продаж игр для различных платформ по регионам, млн. шт.')
for i, region in enumerate(regions):
    for j, sale in enumerate(sales):
        df_portrets.groupby('platform')[sale].sum().plot(kind='bar', ax=axs[j])
```

### Проверка гипотез о средних пользовательских рейтингах 
[В начало](#intro)

Проверим гипотезы о том, что:
- средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
- средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

#### Гипотеза о равенстве средних пользовательских рейтингов платформ Xbox One и PC для периода 2014...2016гг.
[В начало](#intro)

**Формулируем нулевую гипотезу H₀:**<br/>
**Разницы в средних пользовательских рейтингах платформ "Xbox One" и "PC" нет, исследуемые параметры равны.**<br/>
То есть **H₀: rating_xone = rating_pc**.

**Формулируем альтернативную гипотезу H₁:**<br/>
**Cредний пользовательский рейтинг платформы "Xbox One" не равен среднему пользовательскому рейтингу платформы "PC".**<br/> 
То есть **H₁: rating_xone != rating_pc**.<br/>
*(двухсторонняя гипотеза)*</br>

Фактически мы проверяем гипотезу о равенстве средних двух генеральных совокупностей по взятым из них выборкам.
    
**Устанавливаем уровень значимости alfa = 0.05** (пороговый уровень статистической значимости alfa или ошибка первого рода, то есть вероятность ошибочно отклонить нулевую гипотезу составляет 5 из 100).

**Для выполнения расчетов исключим отсутствующие значения в оценках пользователей**.

```python
# Проверяем нулевую гипотезу H₀:
# H₀: rating_xone = rating_pc

# готовим выборки по оценкам пользователей для платформ "Xbox One" и "PC":
rating_xone = df_2014_2016.query('platform == "XOne"')['user_score']
rating_pc = df_2014_2016.query('platform == "PC"')['user_score']
print('Количество игр на платформе:')
print('  - "Xbox One":', (rating_xone.shape[0]),'шт.')
print('  - "PC":', (rating_pc.shape[0]),'шт.')
print()
print('Количество игр, которым пользователи не дали оценку, на платформе:')
print('  - "Xbox One":', round(rating_xone.isna().sum()), 'шт. или ', round((rating_xone.isna().sum() / rating_xone.shape[0]) * 100), ', % от всех игр платформы.')
print('  - "PC":', round(rating_pc.isna().sum()), 'шт. или ', round((rating_pc.isna().sum() / rating_pc.shape[0]) * 100), ', % от всех игр платформы.')
print()
print('Средняя оценка пользователей для игр на платформе:')
print('  - "Xbox One":', round(rating_xone.mean()),'из 10,0 возможных баллов.')
print('  - "PC":', round(rating_pc.mean()),'из 10,0 возможных баллов.')
print()
# для выполнения расчетов исключаем отсутсвующие значения:
rating_xone = rating_xone.dropna()
rating_pc = rating_pc.dropna()

# устанавливаем уровень статистической значимости
# (если p-value окажется меньше него - отвергнем гипотезу):
alpha = .05

# проверяем гипотезу о равенстве среднего двух генеральных совокупностей по взятым из них выборкам
# с помощью статистического теста с использованием t-распределения: t-теста Стьдента
# (поскольку выборки разного размера и/или можно предположить, что дисперсии у генеральных совокупностей разные, то
# при вызове метода scipy.stats.ttest_ind() указываем equal_var = False):
results = st.ttest_ind(rating_xone, 
                       rating_pc, 
                       equal_var = False)

# находим p-значение, то есть вероятность получить наблюдаемый или еще более далекий от предполагаемого результат при условии, 
# что нулевая гипотеза верна:
print('p-значение:', results.pvalue)
print('уровень статистической значимости:', alpha)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

После проверки: **нулевая гипотеза H₀ о том, что разницы в средних оценках игр от пользователей для платформ "Xbox One" и "PC" нет и исследуемые параметры равны (H₀: rating_xone = rating_pc) - не была отвергнута**.
Другими словами по имеющейся выборке возможно считать, что в среднем оценки для этих двух платформ одинаковые. 

Полученное значение p-value = 0,1160 говорит о том, что хотя средние оценки платформ "Xbox One" и "PC" отличаются (7 и 6 баллов из 10 возможных), <b>с вероятностью 11,6% такое или большее различие можно получить случайно. </b> Это слишком большая вероятность (при принятом уровне статистической значимости alpha = 5%), чтобы делать вывод о значимом различии между средними оценками платформ.

#### Гипотеза о неравенстве средних пользовательских рейтингов жанров Action и Sports для периода 2014...2016гг.
[В начало](#intro)

**Формулируем нулевую гипотезу H₀:**<br/>
**Разницы в средних оценках игр от пользователей для жанров Action и Sports нет, исследуемые параметры равны.**<br/>
То есть **H₀: rating_action = rating_sports**.

**Формулируем альтернативную гипотезу H₁:**<br/>
**Средние оценки игр от пользователей для жанра Action выше, чем для жанра Sports.**<br/> 
То есть **H₁: rating_action > rating_sports**.<br/>
*(односторонняя гипотеза вправо: выборочное среднее для rating_action значимо больше **предполагаемого среднего для rating_sports**)*

**Устанавливаем уровень значимости alfa = 0.05** (пороговый уровень статистической значимости alfa или ошибка первого рода, то есть вероятность ошибочно отклонить нулевую гипотезу составляет 5 из 100).

**Для выполнения расчетов исключим отсутствующие значения в оценках пользователей**.

```python
# Проверяем нулевую гипотезу H₀:
# H₀: rating_action = rating_sports

# готовим выборки по оценкам пользователей для жанров "Action" и "Sports":
rating_action = df_2014_2016.query('genre == "Action"')['user_score']
rating_sports = df_2014_2016.query('genre == "Sports"')['user_score']
print('Количество игр в жанрах:')
print('  - "Action":', (rating_action.shape[0]),'шт.')
print('  - "Sports":', (rating_sports.shape[0]),'шт.')
print()
print('Количество игр, которым пользователи не дали оценку, для жанров:')
print('  - "Action":', round(rating_action.isna().sum()), 'шт. или ', round((rating_action.isna().sum() / rating_action.shape[0]) * 100), ', % от всех игр платформы.')
print('  - "Sports":', round(rating_sports.isna().sum()), 'шт. или ', round((rating_sports.isna().sum() / rating_sports.shape[0]) * 100), ', % от всех игр платформы.')
print()
print('Средняя оценка пользователей для игр жанров:')
print('  - "Action":', round(rating_action.mean()),'из 10,0 возможных баллов.')
print('  - "Sports":', round(rating_sports.mean()),'из 10,0 возможных баллов.')
print()
# для выполнения расчетов исключаем отсутсвующие значения:
rating_action = rating_action.dropna()
rating_sports = rating_sports.dropna()

# устанавливаем уровень статистической значимости
# (если p-value окажется меньше него - отвергнем гипотезу):
alpha = .05

# устанавливаем значение с которым проводим сравнение по жанру "Sports":
rating_s = round(rating_sports.mean())
print('принимаем значение предполагаемого среднего (с которым проводим сравнение) равным среднему по жанру "Sports":', rating_s) 

# проверяем гипотезу с помощью статистического теста с использованием t-распределения: t-теста Стьдента
# через метод scipy.stats.ttest_1samp(), проверяющий двустороннюю гипотезу. Поскольку нужен дносторонний тест,
# выполняем двусторонний, а выданное им p-value поделим пополам - так мы получим одностороннюю вероятность
# отклонения наблюдаемого значения от предполагаемого:
results = st.ttest_1samp(
                         rating_action, 
                         rating_s, 
                         )

# находим p-значение, то есть вероятность получить наблюдаемый или еще более далекий от предполагаемого результат при условии, 
# что нулевая гипотеза верна:
print('p-значение:', results.pvalue / 2) # делим на 2, поскольку альтернативная гипотеза H₁ - односторонняя
print('уровень статистической значимости:', alpha)

# делим на 2, поскольку альтернативная гипотеза H₁ - односторонняя, кроме того
# добавляем дополнительное условие сравнения средних ">", поскольку альтернативная гипотеза односторонняя вправо:
if (results.pvalue / 2) < alpha and (rating_action.mean() > rating_s):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

После проверки: нулевая гипотеза H₀ о том, что разницы в средних оценках игр от пользователей для жанров "Action" и "Sports" нет и исследуемые параметры равны (H₀: rating_action = rating_sports) - **была отвергнута**. Альтернативная гипотеза: средние оценки игр от пользователей для жанра Action выше, чем для жанра Sports.<br/>
Иными словами полученное значение p-value ~ 0% говорит о том, что зафиксированное или большее различие в средних оценках жанров "Action" и "Sports" (7 и 5 баллов из 10 возможных) с вероятностью почти 0% можно получить случайно (т.е. почти невозможно случайно получить). Это слишком маленькая вероятность (при принятом уровне статистической значимости alpha = 5%), на основании которой с учетом подтверженного неравенства "выборочное среднее для rating_action больше предполагаемого среднего для rating_sports" возможно сделать вывод о значимом различии между средними оценками жанров (т.е. отклонении нулевой гипотезы в пользу альтернативной).

Таким образом, **самыми востребованными играми оказались игры жанров Action, Shooter, Sports, Role-Playing** (наибольшее количество релизов и проданных копий, наибольшие медианы проданных копий и/или более длинные ряды успешно проданных игр).

**Составление портрета пользователя для регионов:<br/>
Северная Америка (NA), Европейский союз (EU), Япония (JP), остальные регионы (Other).**<br/>

**Наиболее востребованные игроками платформы.**<br/>
В Северной Америке, Европейском союзе и в других регионах **первое место по долям продаж занимают игры для платформы PS4** (от 34,7% до 54,5% в зависимости от региона), на втором месте - XOne (15,3%...28,6%), на третьем и четвертом местах либо PS3, либо Х360 (Северная Америка). Особняком выступает составляет Япония, где на первом месте платформа 3DS (47,5%), на втором и третьем - PS4 и PS3 соответственно.<br/> 
Также следует отметить, что продажи игр для платформ, не вошедших в ТОР-5, для Японии составляют значительную долю рынка (23,9%), а для остальных регионов эта доля 13,8% и менее.<br/>
Выявленные отличия между примерно одинаковой картиной распространенности игровых платформ во всех регионах за исключением Японии, вероятно, объясняются особым рынком видеоигр в этой стране: индустрия в стране восходящего солнца во многом шла своим собственном путем и влияние продуктов из США не так сильно как на других территориях.

**Наиболее востребованные игроками игровые жанры.**<br/>
В Северной Америке, Европейском союзе и в других регионах практически одинаковые позиции - **первое или с минимальным отрывом второе место - по долям продаж занимают игры жанров 'Action' и 'Shooter'** (от 24,2% до 27,8% в зависимости от региона и жанра), на третьем месте - игры жанра 'Sports'. Предпочтания игроков из Японии отличаются: на первом месте жанр 'Role-Playing' (33,5%), на втором - 'Action' (31,8%), а доля игр 'Shooter' (5,2%) совсем невелика по сравнению с другими регионами.<br/>
Для Японии присуща еще одна характерная особенность: продажи игр для жанров, не вошедших в ТОР-5, составляют значительную долю рынка (20,0%), а для остальных регионов эта доля 16,2% и менее.<br/>
Как и для случая с наиболее распространными игровыми платформами, выявленные отличия между распространненостью жанров игр во всех регионах и Японии, вероятно, объясняются особым рынком видеоигр и менталитетом японских игроков: индустрия здесь во многом развивалась самостоятельным путем.

**Влияние рейтинга ESRB на покупку игр.**<br/>
В отношении доли продаж игр в зависимости от присвоенного ассоциацией ESRB рейтинга картина для Северной Америки, Европейского союза и в других регионах, за исключением Японии, фактически повторяют друг друга: **больше всего продано игр с рейтином 'М'** (33,9...35,1%). Далее идут игры с рейтином 'Е', 'Т', 'Е10+'. Доля игр без рейтинга колеблится в районе 21,8...22,8%.<br/>
Что касается Японии, то здесь доля продаж игр без рейтинга составляет 61,1%, а первое место среди рейтниговых игр занимают игры рейтинга 'Т' (15,9%).<br/>
Как было отмечено выше, выявленные отличия между распространненостью игр в зависимости от рейтинга ESRB во всех регионах и Японии, по всей видимости, объясняются особым рынком видеоигр и менталитетом японских игроков.
